**4.2.Multi Linear Regression Model**

Dropping Cities and Flight Name column as these are not actionable

Created a new data frame by dropping several columns from the original data frame

```python
df1= df.copy()
```

```python
df1.drop(['City 1'], axis =1 , inplace =True)
df1.drop(['City2'], axis = 1, inplace= True)
df1.drop(['Market Leading Airline(MLA)'], axis = 1, inplace=True)
df1.drop(['Low Price Airline(LPA)'], axis =1, inplace = True)
```

```python
df1.info()
```

**4.2.1.Splitting Data into Training and Testing**

```python
y= df1.pop('Average flight fare')
x= df1
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.3, random_state=1)
```

**4.2.2. Rescaling the Features**  Rescaling featurs is a common preprocessing step in machine learning to ensure that all features have the same scale.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Applying scaler() to all columns except the dummy 'variable'
vars =['Distance', 'Average Weekly Passengers','Market Share MLA','Average MLA fare','Market Share LPA','Average LPA Fare']
x_train[vars] = scaler.fit_transform(x_train[vars])
x_test[vars] = scaler.fit_transform(x_test[vars])
```

```python
x_train.head()
```

**4.2.3. Fitting Regression Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()    # call the function
model.fit(x_train, y_train)   # fit the model
print(model.intercept_)
print(model.coef_)
```

```python
#See the error parameters
from sklearn import metrics

# make prediction

pred = model.predict(x_test)
print(metrics.mean_absolute_error(y_test, pred))
print(metrics.mean_absolute_error(y_test, pred))

# Evolute the model
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```

**4.2.4.Residual Analysis** is the difference between the observed values and the predicted values of the dependent variable

```python
y_train_price = model.predict(x_train)
```

```python
# plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train- y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)
plt.xlabel('Erros', fontsize =18)
```

Error terms are normally distributed - hence prediction can be made

**4.2.5. Making Prediction using final model**

```python
# plotting y_test and pred to understand the spread
fig = plt.figure()
plt.scatter(y_test, pred)
fig.suptitle('y_test vs pred', fontsize =30)
plt.xlabel ('y_test',fontsize =18)
plt.ylabel('pred', fontsize = 16)
```

```python
from sklearn import metrics
metrics. explained_variance_score(y_test, pred)
```

Model is able to explain 88 % of variance of test data
